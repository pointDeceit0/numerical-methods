# Pseudo-random numbers



## Method 1

$x_0$ - arbitary number

```python
x0 = 1234567
n = 10

for _ in range(n):
    x0 **= 2
    
    x0 = x0 // 10e2

    x0 = (x0 / 10e6 - x0 // 10e6) * 10e6

    print(x0)
```

## Method 2

$x_n = (ax_{n-1} + c) \; mod \; d$

On some values it could seemed that it is random, but it is not so

```python
def meth2(a, c, d, n=1_000_000, x=1) -> float:
    """
    Generates pseudo-random number and plots it
    """
    ox, oy, oz = [], [], []

    for i in range(1, n):
        x = (a * x + c) % d
        if i % 3 == 0:
            ox.append(x / d)
        elif i % 3 == 1:
            oy.append(x / d) 
        else:
            oz.append(x / d)

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    ax.scatter(ox, oy, oz, s=0.2)

    plt.show()
```

```python
a, c, d = 2**16 + 3, 0, 2**31 + 7
meth2(a, c, d)
```

__non-randomness situation:__

```python
a, c, d = 2**16 + 3, 0, 2**31
meth2(a, c, d)
```

## Random Numbers in Interval

Generator of random number in $[0, n]$ interval

$U[0, 1] * n + 1$

```python
n = 3290
np.floor(random.random() * n) + 1
```

__In $[-n, 0]$ interval:__

```python
np.floor(random.random() * n) + 1 - n
```

__In $[0, n]$ with $frac$ fractional part:__

```python
frac = 10
(np.floor(random.random() * n * frac) + 1) / frac
```
