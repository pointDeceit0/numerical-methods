# The Secretary Problem

[wiki](https://en.wikipedia.org/wiki/Secretary_problem)

The solution strategy we take minimum (1 - the best) from first _strategy_ secretaries and then we begin to compare him with another \[strategy;n].

$\lim_{x \to \infty} \frac{n}{secretary_{max\_prob}} = e$

```python
n = 1000 # number of secretaries/brides
h = 500 # times of attempts

probs = []

for strategy in range(1, n + 1):
    cur = []
    c = 0
    for _ in range(h):
        a = np.random.permutation(range(1, n + 1))
            
        m = min(a[:strategy])
        for i, v in enumerate(a[strategy:]):
            if v < m:
                if v == 1:
                    c += 1
                break
            if a[strategy + i] < m: m = a[strategy + i]
    probs.append(c / h)

print(n / probs.index(max(probs)))

# creating
fig = plt.figure()
ax = fig.add_subplot()
# grid step
ax.set_xticks(np.arange(0, n, 1))
ax.grid("both")
# plot
ax.plot(range(1, n + 1), probs)
plt.show()
```
